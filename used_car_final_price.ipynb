# Split dataset for train and test

```python
X = df1.drop('Price',axis=1)
y = df1['Price']
print('shape of x = ', X.shape)
print('shape of y= ',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=51)
print('shape of X_train = ',X_train.shape)
print('shape of y_train = ',y_train.shape)
print('shape of X_test = ',X_test.shape)
print('shape of y_test = ',y_test.shape)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

# Model Building and training

# Xgboost

```python
from sklearn.metrics import mean_squared_error
def rmse(y_test, y_pred):
  return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
import xgboost
xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train,y_train)
xgb_reg_score=xgb_reg.score(X_test,y_test) 
xgb_reg_rmse = rmse(y_test, xgb_reg.predict(X_test))
xgb_reg_score, xgb_reg_rmse
```

# RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
rfr_score=rfr.score(X_test,y_test) 
rfr_rmse = rmse(y_test, rfr.predict(X_test))
rfr_score, rfr_rmse
```

# ExtraTree Regressor

```python
from sklearn.ensemble import ExtraTreesRegressor
modelextra = ExtraTreesRegressor(n_estimators=100,verbose=1,n_jobs=4,warm_start=True,max_features=150, min_samples_split=2)
modelextra.fit(X_train, y_train)
```

```python
model_score=modelextra.score(X_test,y_test) 
model_rmse = rmse(y_test, modelextra.predict(X_test))
model_score, model_rmse
```

```python
print(pd.DataFrame([
              {'Model': 'Random Forest','Score':rfr_score, "RMSE":rfr_rmse},
              {'Model': 'ExtraTree Rgressor','Score':model_score, "RMSE":model_rmse},
              {'Model': 'XGBoost','Score':xgb_reg_score, "RMSE":xgb_reg_rmse}],
             columns=['Model','Score','RMSE']))
```
